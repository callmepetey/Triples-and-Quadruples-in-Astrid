# Tuple Plots

## Initializations

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
# you need to install BigFile
from bigfile import BigFile
import glob,os,struct
# from astropy.cosmology import FlatLambdaCDM
import astropy.units as u
from scipy.interpolate import interp1d
from colossus.cosmology import cosmology

import seaborn as sns

sns.set()
sns.set_palette("Set2")
sns.set_style('ticks',{'ytick.direction':'in','xtick.direction':'in'})

cmap = plt.get_cmap("Set2")
sns.set_context("paper", font_scale=1.7,rc={"axes.linewidth": 1.3,"lines.linewidth": 2.5,"patch.linewidth": 2.2})
from matplotlib import rcParams as rc
import pickle
import warnings

import import_ipynb
import Tuples_Information as tuples_z2_z3
```

```python
triples_z2_dr30 = tuples_z2_z3.trips_info_z2_dr30
triples_z3_dr30 = tuples_z2_z3.trips_info_z3_dr30
triples_z2_dr200 = tuples_z2_z3.trips_info_z2_dr200
triples_z3_dr200 = tuples_z2_z3.trips_info_z3_dr200

quadruples_z2_dr30 = tuples_z2_z3.quads_info_z2_dr30
quadruples_z3_dr30 = tuples_z2_z3.quads_info_z3_dr30
quadruples_z2_dr200 = tuples_z2_z3.quads_info_z2_dr200
quadruples_z3_dr200 = tuples_z2_z3.quads_info_z3_dr200
```

```python
triples_dr30 = np.concatenate((triples_z3_dr30, triples_z2_dr30))
quadruples_dr30 = np.concatenate((quadruples_z3_dr30, quadruples_z2_dr30))

triples_dr200 = np.concatenate((triples_z3_dr200, triples_z2_dr200))
quadruples_dr200 = np.concatenate((quadruples_z3_dr200, quadruples_z2_dr200))

tuples_dr30 = np.concatenate((triples_dr30, quadruples_dr30))
tuples_dr200 = np.concatenate((triples_dr200, quadruples_dr200))
```

## BH Mass vs LumX

```python
fig, ax = plt.subplots(figsize = (9, 6))

ax.scatter(tuples_dr200['BHMass'],tuples_dr200['LumX'],s=100,marker='x',c='blue',label=r'$\Delta r_{200}\ {\rm tuples}$')
ax.scatter(tuples_dr30['BHMass'],tuples_dr30['LumX'],s=100,marker='x',c='red',label=r'$\Delta r_{30}\ {\rm tuples}$')

ax.axhline(y=10**42,linestyle='--',color='grey')

ax.set_xlabel(r'$M_{\rm BH}[M_\odot]$',fontsize=20)
ax.set_ylabel(r'$L_{\rm x}[erg/s]$',fontsize=20)
ax.set_xscale('log')
ax.set_yscale('log')
ax.legend(loc='best')

plt.savefig("tuples_bhmass_lumx",dpi=150)

plt.show()
```

### Investigating $z=2$ and $z=3$ Plots

```python
fig, ax = plt.subplots(figsize = (9, 6))

ax.scatter(triples_z2_dr30['BHMass'],triples_z2_dr30['LumX'],s=100,marker='x',c='blue',
           label=r'$z=2\ \Delta r_{30}\ {\rm triples}$')
ax.scatter(triples_z2_dr200['BHMass'],triples_z2_dr200['LumX'],s=100,marker='x',c='red',
           label=r'$z=2\ \Delta r_{200}\ {\rm triples}$')
ax.scatter(quadruples_z2_dr30['BHMass'],quadruples_z2_dr30['LumX'],s=100,marker='x',c='green',
           label=r'$z=2\ \Delta r_{30}\ {\rm quadruples}$')
ax.scatter(quadruples_z2_dr200['BHMass'],quadruples_z2_dr200['LumX'],s=100,marker='x',c='purple',
           label=r'$z=2\ \Delta r_{200}\ {\rm quadruples}$')

ax.axhline(y=10**42,linestyle='--',color='grey')

ax.set_xlabel(r'$M_{\rm BH}[M_\odot]$',fontsize=20)
ax.set_ylabel(r'$L_{\rm x}[erg/s]$',fontsize=20)
ax.set_xscale('log')
ax.set_yscale('log')
ax.legend(loc='best')
```
